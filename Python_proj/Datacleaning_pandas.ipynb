```python
import pandas as pd
```

```python
df = pd.read_csv("messy_sales_big.csv")
print(df.to_string())
```

```python
# remove duplicates
df[df.duplicated(keep = False)] 
print(df.to_string())
```

```python
# remove the orderID of NAN 
df = df[df['OrderID'].notna()]
print(df.to_string())
```

```python
#convert OrderID datatype to number
df['OrderID'] = df['OrderID'].astype(int)
print(df.to_string())
```

```python
#replace NaN of customerName
df['CustomerName'] = df['CustomerName'].fillna('Customer_2')
print(df.to_string())
```

```python
# define the default product based on category
def prod_def(val):
    if pd.isna(val['Product']):
        if val['Category'] == 'Dairy':
            return 'Cheese'
        elif val['Category'] == 'Bakery':
            return 'Milk'
        elif val['Category'] == 'Produce':
            return 'Bread'
        elif val['Category'] == 'Seafood':
            return 'Salmon'
        elif  val['Category'] == 'Beverages':
            return 'Cola'
        else:
            return 'Banana'
    else:
        return val['Product']

df['Product'] = df.apply(lambda x:prod_def(x),axis = 1)
print(df.to_string())
```

```python
# define a default for category
def cat_def(val):
    if pd.isna(val['Category']):
        if val['Product'] in 'Milk':
            return 'Bakery'
        elif val['Product'] == 'Cheese':
            return 'Dairy'
        elif val['Product'] == 'Cola':
            return 'Beverages'
        elif val['Product'] == ['Salmon','Halibut Fillet','Cod Fillet']:
            return 'Seafood'
        elif val['Product'] == ['Bread','Orange Juice']:
            return 'Produce'
        else:
            return 'Fruits'
    else:
        return val['Category']
df['Category'] = df.apply(lambda x:cat_def(x),axis = 1)
print(df.to_string())
```

```python
# fill na of quanitity with median values
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
print(df.to_string())
```

```python
# fill na of Unit Price with mode values
df['UnitPrice'] = df['UnitPrice'].fillna(df['UnitPrice'].mean())
print(df.to_string())
```

```python
# fill na of OrderDate with max_date values
df['OrderDate'] = df['OrderDate'].fillna(df['OrderDate'].max())
print(df.to_string())
```

```python
# fill na of Country with default values
df['Country'] = df['Country'].fillna('USA')
print(df.to_string())
```

```python
#add total sales
df['Total'] = df['Quantity'] * df['UnitPrice']
print(df.to_string())
```

```python
# group products by category and find the avg,total
totals = df.groupby('Category').agg(totalCost = ('Total','sum'), AvgCost = ('Total','mean'))
print(totals.to_string())
```
