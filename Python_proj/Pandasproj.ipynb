```python
import pandas as pd
```

```python
df = pd.read_csv("sales_data_portfolio.csv")
print(df.to_string())
```

```python
#Show the summary info (.info()) and descriptive statistics (.describe()).
print(df.info())
```

```python
print(df.describe())
```

```python
#Select only the CustomerName and Country columns.
print(df[['CustomerName','Country']])
```

```python
# add new column total to find the total price 
df['Total'] = df['Quantity'] * df['UnitPrice']
print(df.to_string())
```

```python
#Filter records with category = seafood and country == UK
seafood_data = df[(df['Category'] == 'Seafood') & (df['Country'] == 'UK')]

print(seafood_data)
```

```python
# find all orders where Quantity is greater than 5
quangreater_5 = df[df['Quantity'] > 5]
print(quangreater_5)
```

```python
# get the top 5 totals 
res = df.nlargest(n = 5,columns = 'Total')
print(res)
```

```python
#find the total quantity sold for each product
group = df.groupby("Product")
quan_sold = group['Quantity'].sum()
#compute the average unitprice for each product
unit_price = group['UnitPrice'].mean()
print(quan_sold,unit_price)
```

```python
# count the number of orders from each country
group = df.groupby("Country")
num_of_orders = group['OrderID'].count()
sorted_res = num_of_orders.nlargest()
print(num_of_orders,sorted_res)
```

```python
# filter order in the last 6 months 
df['OrderDate'] = pd.to_datetime(df['OrderDate'], format='%Y-%m-%d')
last_6months = df[(df['OrderDate'] <= '2026-02-09') & (df['OrderDate'] >= '2025-08-01')]
                                 
print(last_6months.tail())
```

```python
# Group by Country and Category, then calculate total revenue for each group
vals = df.groupby(['Country','Category'])['Total'].sum()
print(vals)

```

```python
#Find the most popular product in terms of quantity sold
popular_prod = df.groupby("Product")['Quantity'].count().nlargest()
print(popular_prod)
```

```python
#Sort all orders by revenue in descending order
sorteddf = df.sort_values(by='Total',ascending= False)
print(sorteddf.head())
```

```python
#Find customers who ordered more than 2 different product categories.
vals = df.groupby("CustomerName")["Category"].nunique()
res = vals[vals > 2]
print(res)

```

```python
#Calculate the monthly total revenue and visualize it using a line plot.
import matplotlib
```

```python


df['month'] = pd.to_datetime(df['OrderDate'],format='%m').dt.month_name()

gp = df.groupby('month')['Total'].sum()
plt = gp.plot(kind = 'line')

print(gp,plt)
```

```python
#Create a summary table showing, for each country:Total orders,Total revenue,Average order value"""
summary = df.groupby('Country').agg(
    Total_orders=('OrderID', 'count'),
    Total_revenue=('Total', 'sum'),
    Avg_order_value=('Total', 'mean')
).reset_index()
print(summary)
# find top 3 prods per total
print(summary.sort_values('Total_revenue',ascending = False).head(3))
```

```python
#Create a pivot table showing Category vs Country with total revenue
vals = df.groupby(['Country','Category']).agg(revenue = ('Total','sum')).reset_index()

print(vals.pivot(index='Country',columns = 'Category',values ='revenue'))
```

```python
#Save a filtered DataFrame with only "Seafood" orders from Poland in 2025 into a new CSV.
seafood_pl = df[(df['Category'] == 'Seafood') & (df['OrderDate'] > '2024-12-31') & (df['Country'] == 'Poland')]
print(seafood_pl)
seafood_pl.to_csv('seafood_pl.csv', sep='\t', encoding='utf-8')
```
